```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Step 1 — fit PCA on dev set only
pca = PCA(random_state=42)
pca.fit(X_dev_scaled)
```

```python
# Step 2 — find optimal number of components (95% variance explained)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Components needed to explain 95% variance: {n_components}")
print(f"Original features : {X_dev_scaled.shape[1]}")
print(f"Reduced features  : {n_components}")
```

```python
# Step 3 — refit PCA with optimal components
pca = PCA(n_components=n_components, random_state=42)

X_dev_pca  = pca.fit_transform(X_dev_scaled)
X_val_pca  = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(f"\nX_dev_pca  : {X_dev_pca.shape}")
print(f"X_val_pca  : {X_val_pca.shape}")
print(f"X_test_pca : {X_test_pca.shape}")
print(f"Variance explained: {pca.explained_variance_ratio_.sum():.4f}")
```
