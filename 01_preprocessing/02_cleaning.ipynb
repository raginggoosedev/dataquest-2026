```python
import polars as pl
import numpy as np
```

```python
df = pl.read_csv("../diabetic_data.csv")
```

# Merge datasets into one

```python
admission_type = pl.read_csv("../admission_type.csv")
admission_source = pl.read_csv("../admission_source.csv")
discharge_disposition = pl.read_csv("../discharge_disposition.csv")
```

```python
df = df.join(admission_type, on="admission_type_id", how="full").rename({"description": "admission_type"}).drop("admission_type_id").drop("admission_type_id_right")
```

```python
df = df.join(admission_source, on="admission_source_id", how="full").rename({"description": "admission_source"}).drop("admission_source_id").drop("admission_source_id_right")
```

```python
df = df.join(discharge_disposition, on="discharge_disposition_id", how="full").rename({"description": "discharge_disposition"}).drop("discharge_disposition_id").drop("discharge_disposition_id_right")
```

```python
df
```

```python
df = df.drop("encounter_id").drop("patient_nbr")
```

```python
df
```

```python
# Delete null rows
df.filter(~pl.any_horizontal(pl.all().is_null()))
```

```python
df = df.filter(~pl.any_horizontal(pl.all().is_null()))
```

```python
df = df.with_columns(
    age = (
        pl.col("age")
        .str.extract_all(r"\d+")  # Extracts all digits as a list of strings
        .list.get(0).cast(pl.Float64) + 
        pl.col("age")
        .str.extract_all(r"\d+")
        .list.get(1).cast(pl.Int64)
    ) / 2
)
```
